```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

Two Sample T-test

```python
df_dragon = df[(df['Type 1']=="Dragon") | (df['Type 2']=="Dragon")]["HP"]
df_grass = df[(df['Type 1']=="Grass") | (df['Type 2']=="Grass")]["HP"]

#Set the hypothesis

#H0: mu_hp dragon <= mu_hp grass
#H1: mu_hp dragon > mu_hp grass

#significance level = 0.05

st.ttest_ind(df_dragon, df_grass, equal_var=False, alternative='greater')
```

Given our pvalue being so small and lower than out significance level of 0.05, we reject the Null Hypothesis 

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
df.columns.unique()
```

```python
# ANOVA Test 
#Set the hypothesis

#H0: mu_attr leg = mu_attr no_leg
#H1: mu_attr leg != mu_attr no_leg

#significance level = 0.05

attributes = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
alpha = 0.05

for attribute in attributes:
    print('\n'f'We are looking at attribute: {attribute}')
    
    df_leg = df[(df["Legendary"]==True)][attribute]
    df_no_leg = df[(df["Legendary"]==False)][attribute]
    p_value = st.ttest_ind(df_leg, df_no_leg,  equal_var=False)[1]
    print(f' p_value = {p_value}')
    if p_value > alpha:
        print("We are not able to reject the null hypothesis")
    else:
        print("We reject the null hypotesis")
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
sch_coor=(-118, 37) #(longitude,latitude)
hos_coor=(-122, 34)

# df['latitude']
# df['longitude']

# # Calculate Euclidean Distance
def school_eu_dis(df, sch_coor):
    s_eu_dis= np.sqrt((df['longitude']-sch_coor[0])**2 + (df['latitude']-sch_coor[1])**2)
    return s_eu_dis

distance1 = round(school_eu_dis(df, sch_coor),2)
df['dist2school']= distance1

def hospi_eu_dis(df, hos_coor):
    h_eu_dis= np.sqrt((df['longitude']-hos_coor[0])**2 + (df['latitude']-hos_coor[1])**2)
    return h_eu_dis

distance2 = round(hospi_eu_dis(df, hos_coor),2)
df['dist2hospi']= distance2

df
```

```python
df['dist2school']
```

```python
print(df.dist2school.describe())  #just to check the min and max of each distance
print(df.dist2hospi.describe())
```

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

```python
df_closeby=df[(df['dist2school']<= 0.50) | (df['dist2hospi'] <= 0.50)]
df_closeby
```

```python
df_farland=df[(df['dist2school']> 0.50) | (df['dist2hospi'] > 0.50)]
df_farland
```

Hypothesis to test: We posit that houses close to either a school or a hospital are more expensive.

```python
#Set the hypothesis


#H0: mu_median_house_value closeby <= mu_median_house_value farland
#H1: mu_median_house_value closeby > mu_median_house_value farland

#significance level = 0.05

df_closeby=df[(df['dist2school']<= 0.50) | (df['dist2hospi'] <= 0.50)]['median_house_value']
df_farland=df[(df['dist2school']> 0.50) | (df['dist2hospi'] > 0.50)]['median_house_value']

st.ttest_ind(df_closeby, df_farland, equal_var=False, alternative='greater')

pvalue = st.ttest_ind(df_closeby, df_farland, equal_var=False, alternative='greater')[1]
print('\n',f'pvalue: {pvalue}')
```

With the current p_value, we fail to reject the null hypothesis, therefore the distance to school or hospital doesn't necessarily have an impact in the prices.
